```python
#Importing Pandas library for cleaning up raw data
import pandas as pd
```

```python
#File name of raw csv, which should be in the same folder/directory as the Preprocessing Notebook
raw_filename = 'NCWIT_DataV2_RawData.csv'
```

```python
rawdata_dataframe = pd.read_csv(raw_filename)
rawdata_dataframe
```

```python
def getColumnData(colname):
    return rawdata_dataframe[colname]

#example
getColumnData('Record#')
```

```python
#Filter Column by string and remove NaN (null or Not a Number values in data)

def filterColByStr(colname,string_filter):
    return rawdata_dataframe[rawdata_dataframe[colname].str.contains(string_filter, na = False )]

#Step-by-step explanation (from inner to outer):
#df['column name'] selects the ids column of the data frame (techincally, the object df[ids] is of type pandas.Series)
#df['ids'].str allows us to apply vectorized string methods (e.g., lower, contains) to the Series
#df['ids'].str.contains('') checks each element of the Series as to whether the element value has the string filter as a substring. The result is a Series of Booleans indicating True or False about the existence of a 'ball' substring.
#df[df['ids'].str.contains('ball')] applies the Boolean 'mask' to the dataframe and returns a view containing appropriate records.
#na = False removes NA / NaN values from consideration; otherwise a ValueError may be returned.
 
filterColByStr('What degrees does your institution offer?',"Bachelor's and master's degrees only")
```

```python
#Filter Multiple columns based on numbers or strings
#searching for integer 456.0 in "Institution" column AND '2005-2006' in "School Year" Column
rawdata_dataframe[rawdata_dataframe["Institution"].isin([456.0]) & rawdata_dataframe["School Year"].isin(['2005-2006','2006-2007'])]
```

```python
#searching for integers 456.0 , 72.0 in "Institution" column AND '2005-2006' in "School Year" Column
rawdata_dataframe[rawdata_dataframe["Institution"].isin([456.0,72.0]) &rawdata_dataframe["School Year"].isin(['2005-2006'])]
```

```python
#Output results to csv

#Filtered for Institutions 456,72.0 for years 2005-2006
cleaned_data_filter1 = rawdata_dataframe[rawdata_dataframe["Institution"].isin([456.0,72.0]) &rawdata_dataframe["School Year"].isin(['2005-2006'])]

#Filtered for Records# > 3000.0 and for years 2005-2006
cleaned_data_filter2 = rawdata_dataframe[  (rawdata_dataframe["Record#"] > 3000.0)  & rawdata_dataframe["School Year"].isin(['2005-2006'])]

print(cleaned_data_filter2["School Year"])

new_dataframe =  pd.DataFrame()
new_dataframe = new_dataframe.append(cleaned_data_filter1)
new_dataframe = new_dataframe.append(cleaned_data_filter2)

new_dataframe.to_csv("clean_data.csv", sep='\t')
```
