```python
#Importing Pandas library for cleaning up raw data
import pandas as pd
import numpy as np
```

```python
#File name of raw csv, which should be in the same folder/directory as the Preprocessing Notebook
raw_filename = 'NCWIT_DataV2_RawData.csv'
```

```python
rawdata_dataframe = pd.read_csv(raw_filename)
rawdata_dataframe
```

## Filter raw data by Program of NCIW Extention Services, Academic Alliances, 2007-2016, Computer Science Program

```python
#Column Name = NCWIT Participant
targetFilter = 'Extension Services, Academic Alliance'
yearFilter =['2007-2008','2008-2009','2009-2010','2010-2011','2011-2012','2012-2013','2013-2014','2014-2015','2015-2016']
new_dataframe = rawdata_dataframe[rawdata_dataframe["NCWIT Participant"].isin([targetFilter]) & rawdata_dataframe["Major Program Name"].isin(["Computer Science"]) & rawdata_dataframe["School Year"].isin(yearFilter)]
```

### Calculate Male,Female,Total Enrollment

```python
new_dataframe['Male Enrollment'] = new_dataframe['Totals, Male: Enrolled in DIFFERENT MAJOR (Tot. M)'] + new_dataframe['Totals, Male: Enrolled in SAME MAJOR (Tot. M)'] - new_dataframe['Totals, Male: Graduated (Tot. M)'] - new_dataframe['Enroll, Male: Transfer Students (Enrl M)'] - new_dataframe['Enroll, Male: New Enrollments (Enrl M)']
new_dataframe['Female Enrollment'] = new_dataframe['Totals, Female: Enrolled in DIFFERENT MAJOR (Tot. F)'] + new_dataframe['Totals, Female: Enrolled in SAME MAJOR (Tot. F)'] - new_dataframe['Totals, Female: Graduated (Tot. F)'] - new_dataframe['Enroll, Female: Transfer Students (Enrl F)'] - new_dataframe['Enroll, Female: New Enrollments (Enrl F)']
new_dataframe['Total Enrollment'] = new_dataframe['Female Enrollment'] + new_dataframe['Male Enrollment']
new_dataframe[['Female Enrollment','Male Enrollment','Total Enrollment']]
```

```python
#Get all Institutions
institutions = new_dataframe.Institution.unique()

prevYear = None
prevTotEnrollment = None
prevMaleEnrollment = None
prevFemaleEnrollment = None

#create empty/NaN for Retention Calc
new_dataframe['Overall Retention'] = np.nan
new_dataframe['Male Retention'] = np.nan
new_dataframe['Female Retention'] = np.nan

for inst in institutions:    
    #get rows of data for each inst
    for i,row in new_dataframe.loc[new_dataframe['Institution'].isin([inst]),:].iterrows():        
        if prevYear is None:
           #Set prevYear for the first time
           prevYear = int(new_dataframe.loc[i]["School Year"][:4])
            
            
           prevTotEnrollment = new_dataframe.loc[i]["Total Enrollment"]
           prevMaleEnrollment = new_dataframe.loc[i]["Male Enrollment"]
           prevFemaleEnrollment = new_dataframe.loc[i]["Female Enrollment"]      
        elif ( int(prevYear + 1) == int(new_dataframe.loc[i]["School Year"][:4] )):
            curYear = int(new_dataframe.loc[i]["School Year"][:4])
            currTotEnrollment =  new_dataframe.loc[i]["Total Enrollment"]
            currMaleEnrollment =  new_dataframe.loc[i]["Male Enrollment"]
            currFemaleEnrollment =  new_dataframe.loc[i]["Female Enrollment"]

            
            
            #calculate retention for consecutive year
            new_dataframe.at[i, 'Overall Retention'] = currTotEnrollment / prevTotEnrollment
            new_dataframe.at[i, 'Male Retention'] = currMaleEnrollment / prevMaleEnrollment
            new_dataframe.at[i, 'Female Retention'] = currFemaleEnrollment / prevFemaleEnrollment
           
            
            #update previous year 
            prevYear = curYear
            prevTotEnrollment = currTotEnrollment
            prevMaleEnrollment = currMaleEnrollment   
            prevFemaleEnrollment = currFemaleEnrollment   
        else:
            prevYear =  int(new_dataframe.loc[i]["School Year"][:4])
            prevTotEnrollment = new_dataframe.loc[i]["Total Enrollment"]
            prevMaleEnrollment = new_dataframe.loc[i]["Male Enrollment"]
            prevFemaleEnrollment = new_dataframe.loc[i]["Female Enrollment"]   
        
    #reset prevYear for next institution
    prevYear = None
    prevTotEnrollment = None
    prevMaleEnrollment = None
    prevFemaleEnrollment = None
new_dataframe[['Female Enrollment','Male Enrollment','Total Enrollment','Female Retention','Male Retention','Overall Retention']][0:266]
```

### Create variables for plotting

```python
# x-axis values
x = ['2007','2008','2009','2010','2011','2012','2013','2014','2015']

#dictionary with the instituition column as the key and the values [0th index is Male Retention, 1th index is Female Retention 2nd index is Overall Retention]
plotDictionary = {}

def checkNaN(raw_val):
    if (pd.isnull(raw_val)):
        return float('nan')
    else:
        return float((1-raw_val)*100)

# create y values for (Male, Female, Overall Retention) for each school
for inst in institutions:
    #set 0 array for start    
    mRetention = 9*[0]
    fRetention = 9*[0]
    oRetention = 9*[0]
    
    #get rows of data for each inst
    for year in x:
        print("searching for year %s") %year
        for i,row in new_dataframe.loc[new_dataframe['Institution'].isin([inst]),:].iterrows():     
            if(int(year) == int(new_dataframe.loc[i]["School Year"][:4] )):
                print(new_dataframe.loc[i,'Institution'])
                if(pd.isnull(new_dataframe.loc[i]["Male Retention"])):
                    mRetention[x.index(year)] = float('nan')
                    fRetention[x.index(year)] = float('nan')
                    oRetention[x.index(year)] = float('nan')
                    break
                else:
                    mRetention[x.index(year)] = checkNaN(new_dataframe.loc[i]["Male Retention"])
                    fRetention[x.index(year)] = checkNaN(new_dataframe.loc[i]["Female Retention"])
                    oRetention[x.index(year)] = checkNaN(new_dataframe.loc[i]["Overall Retention"])
                    break
            #else:
                #print("cant find %s") %year
    #add retention % as value with institution as its key
    plotDictionary[inst] = mRetention, fRetention, oRetention
    
```

## Graphing Data

### Importing libraries

```python
from ipywidgets import interact
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.palettes import Category10
from bokeh.layouts import gridplot
import itertools
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
# Method for cycling through color pallete
def color_gen():
    for c in itertools.cycle(Category10[10]):
        yield c
color = color_gen()
```

```python
# #create a new plot
# plot1 = figure(
#     title="Retention Rates Among CS Students",
#     x_axis_label='Year', y_axis_label='Retention %',
#     y_range=(-5,5)
# )

# plot2 = figure(
#     title="Retention Rates Among CS Students",
#     x_axis_label='Year', y_axis_label='Retention %',
#     y_range=(-5,5)
# )

#draw lines
def drawLines(retentionFilter, inst):
    titleText=""
    if retentionFilter == 0: titleText='(Male)'    
    elif retentionFilter == 1: titleText='(Female)'
    elif retentionFilter == 2: titleText='(Male and Female)'
        
    ## Graph showing all retention rates for all filtered institutions    
    plot1 = figure(
        title="Retention Rates Among CS Students "+titleText,
        x_axis_label='Year', y_axis_label='Retention %',
        y_range=(-500,500)    
    )
    
    for institution in plotDictionary.keys():
        y = plotDictionary[institution][retentionFilter]   
        plot1.line(x, y, legend=str(institution), line_width=2, color=next(color))
    
    ## Graph showing all retention rates for selected institutions
    plot2 = figure(
    title="Retention Rates Among CS Students "+ titleText +" for Institution "+str(inst) ,
        x_axis_label='Year', y_axis_label='Retention %',
        y_range=(-500,500)
    )

    y2 = plotDictionary[inst][retentionFilter]   
    plot2.line(x, y2, legend=str(inst), line_width=2, color=next(color))

    plots = gridplot([[plot1,plot2]])
    show(plots, notebook_handle=True)   
```

```python
#callback method that hanldes graph updateswhen a change is made to the widget 
def update(FilterGender, ViewInstitution):
    if   FilterGender == "Male Retention": 
        drawLines(0, float(inst))
    elif FilterGender == "Female Retention": 
        drawLines(1,float(inst))
    elif FilterGender == "Overall Retention": 
        drawLines(2, float(inst))
    push_notebook()    
```

```python
# Draw Graphs and control Widgets
interact(update, FilterGender=["Male Retention", "Female Retention", "Overall Retention"], ViewInstitution= map(str, institutions))
```
