## Filter raw data by Program of NCIW Extention Services, Academic Alliances

```python
#Column Name = NCWIT Participant
targetFilter = 'Extension Services, Academic Alliance'
yearFilter =['2007-2008','2009-2010','2010-2011','2011-2012','2012-2013','2013-2014','2014-2015','2015-2016']
new_dataframe = rawdata_dataframe[rawdata_dataframe["NCWIT Participant"].isin([targetFilter]) & rawdata_dataframe["School Year"].isin(yearFilter)]

#Masking
#mask = rawdata_dataframe.isin({'NCWIT Participant': [targetFilter], 'School Year': ['2005-2006', '2006-2007', '2007-2008']})
#rawdata_dataframe.where(mask)
```

### Calculate Male,Female,Total Enrollment

```python
new_dataframe['Male Enrollment'] = new_dataframe['Totals, Male: Enrolled in DIFFERENT MAJOR (Tot. M)'] + new_dataframe['Totals, Male: Enrolled in SAME MAJOR (Tot. M)'] - new_dataframe['Totals, Male: Graduated (Tot. M)'] - new_dataframe['Enroll, Male: Transfer Students (Enrl M)'] - new_dataframe['Enroll, Male: New Enrollments (Enrl M)']
new_dataframe['Female Enrollment'] = new_dataframe['Totals, Female: Enrolled in DIFFERENT MAJOR (Tot. F)'] + new_dataframe['Totals, Female: Enrolled in SAME MAJOR (Tot. F)'] - new_dataframe['Totals, Female: Graduated (Tot. F)'] - new_dataframe['Enroll, Female: Transfer Students (Enrl F)'] - new_dataframe['Enroll, Female: New Enrollments (Enrl F)']
new_dataframe['Total Enrollment'] = new_dataframe['Female Enrollment'] + new_dataframe['Male Enrollment']
```

```python
#Filter Column by string and remove NaN (null or Not a Number values in data)

def filterColByStr(colname,string_filter):
    return rawdata_dataframe[rawdata_dataframe[colname].str.contains(string_filter, na = False )]

#Step-by-step explanation (from inner to outer):
#df['column name'] selects the ids column of the data frame (techincally, the object df[ids] is of type pandas.Series)
#df['ids'].str allows us to apply vectorized string methods (e.g., lower, contains) to the Series
#df['ids'].str.contains('') checks each element of the Series as to whether the element value has the string filter as a substring. The result is a Series of Booleans indicating True or False about the existence of a 'ball' substring.
#df[df['ids'].str.contains('ball')] applies the Boolean 'mask' to the dataframe and returns a view containing appropriate records.
#na = False removes NA / NaN values from consideration; otherwise a ValueError may be returned.
 
filterColByStr('What degrees does your institution offer?',"Bachelor's and master's degrees only")
```
